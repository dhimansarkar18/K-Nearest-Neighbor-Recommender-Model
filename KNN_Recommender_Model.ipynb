#III. Data Cleaning and Processing

Defining output variables when "a" and "b" are the input variables

```python
a = movies.drop(["Movie Name","Movie ID","Label"], axis=1)
b = movies["Movie Name"]

neighbor = NearestNeighbors(n_neighbors=5)
neighbor.fit(a,b)
```

#IV. KNN Model (K Nearest Neighbors): Recommendation System

```python
#Movie Reccomendation with respect to the movie "The Post" having (IMDB Rating = 7.2, Biography = Yes, Drama = Yes, Thriller = No, Comedy = No, Crime = No, Mystery = No, History = Yes)

post_movie = [7.2,1,1,0,0,0,0,1]
distance, index = neighbor.kneighbors([post_movie])
print(index)
```

```python
for motion_picture_reccomendations in index:
  print("Based on your preferences you may also like: ")
  print(movies.iloc[motion_picture_reccomendations]["Movie Name"])
```
